```python
#Given the above issue, I needed to convert the strings into integers to remove decimals with code below.

score_columns = [
    "Score 2026",
    "Political Context",
    "Economic Context",
    "Legal Context",
    "Social Context",
    "Safety",
    "Score N-1",
    "Score evolution"
]

for col in score_columns:
    Americas[col] = (
        Americas[col]
        .str.replace(",", ".", regex=False)
        .astype(float)
    )
```

## **Step 5:** Find the average press freedom score by RSF region. This will be used to compare the Americas to the other regions we found earlier in **Step 3** and see and how their standings evolved from 2025 (several cells below). The information from this will also be used to generate a chart to answer - **1) How do the Americas compare with other RSF regions in overall press freedom scores in 2026?** 

```python
#score for all regions and 5 indicators, not just for americas or last df. 
score_columns = [
    "Score 2026",
    "Political Context",
    "Economic Context",
    "Legal Context",
    "Social Context",
    "Safety",
    "Score N-1",
    "Score evolution"
]

for col in score_columns:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(",", ".", regex=False)
        .astype(float)
    )
```

```python
df.info()
```

### **Code below gives numbers for Chart 1:** Average Press Freedom Score by RSF Region (2026)

```python
#the above checked for floats. 
#this cell will now calculate regional avg. for CHART 1.

region_scores = df.groupby("Zone")["Score 2026"].mean().sort_values(ascending=False)

region_scores
```

```python
#Convert it into a DF to change all region names to be in English. Americas in French this time even after last change b/c of old DF.

region_scores_df = region_scores.reset_index()

region_scores_df["Zone"] = region_scores_df["Zone"].replace({
    "Amériques": "Americas",
    "Asie-Pacifique": "Asia-Pacific",
    "Afrique": "Africa",
    "UE Balkans": "Europe/Balkans"
})
```

is fixed in my visuals and the code remains accurate

```python
#check that the above worked

region_scores_df
```

```python
#added this a bit later after realizing/remembering that the CSV includes scores from
#2025, allowing me to find the average for regions based on their coutries' scores in 2025.
region_change = df.groupby("Zone")[["Score N-1", "Score 2026", "Score evolution"]].mean().round(2)
region_change = region_change.sort_values("Score evolution") 

region_change_df = region_change.reset_index() #renamed again bc new df name.
region_change_df["Zone"] = region_change_df["Zone"].replace({
    "Amériques": "Americas",
    "Asie-Pacifique": "Asia-Pacific",
    "Afrique": "Africa",
    "UE Balkans": "Europe/Balkans"
})
region_change_df
```

## **Step 8:** Find the average press freedom indicator scores across the Americas. This answers - **4) Which Press Freedom indicators are weakest across the region?/Which of RSF's five indicators has the lowest average score across the Americas?** This information will also be used for Chart 3.

```python
indicator_cols = [
    "Political Context",
    "Economic Context",
    "Legal Context",
    "Social Context",
    "Safety",
]  

chart3_data = (
    Americas[indicator_cols]
    .mean()# average each column across all countries
    .round(2) #round the numbers by two decimal points like I did earlier to match.
    .reset_index()
)
chart3_data.columns = ["Indicator", "Average score"]
chart3_data = chart3_data.sort_values("Average score")

print(chart3_data)
```

### **Analysis 4:** 


According to RSF 2026 review(https://rsf.org/en/2026-rsf-index-press-freedom-25-year-low?data_type=general&year=2026), "the recent history of press freedom in the Americas has been shaped by the rise in violence perpetrated by two offenders: **organised crime and political actors**." In their text, they explain they use 5 indicators to measure press freedom: **political context, legal framework, economic context, sociocultural context and safety.**

**Political Actors:** Calls out Donald Trump specifically for using his power to influence press in the US and how political actors in foreign countries treat their media. It appears that the US has played a particularly significant role in press freedom in the Americas, with a hand in the decline of scores in individual countries, while losing score itself (7 points). It has also long been looked at as a leader in global press freedom and as a staunch supporter of the 1st amendment, but is now using its international influence to push for regression in press freedom. Specifically mentions Trump's supporters in Latin America (Argentina - difficult, and El Salvador -very serious)

"US President Donald Trump has turned his repeated attacks on the press and journalists into a systematic policy, pushing the US down to 64th place (-7). The detention of Salvadoran journalist Mario Guevara, who was later deported, has contributed to the deterioration of an already tense security environment marked by police violence. The drastic cuts to the US Agency for Global Media (USAGM) workforce had global repercussions, leading to the closure, suspension and downsizing of international broadcasters such as Voice of America (VOA), Radio Free Europe/Radio Liberty (RFE/RL) and Radio Free Asia (RFA) in countries where they were some of the last reliable sources of information."

"Presidents Javier Milei and Nayib Bukele —- Donald Trump’s most vocal supporters in Latin America — have taken their cue from the White House in their approach to the media, with unsurprisingly similar results. Javier Meili’s Argentina (98th, -11) and Nayib Bukele’s El Salvador (143rd, -8) have recorded significant declines, notably linked to the deterioration of these countries’ political and social indicators, reflecting a rise in government hostility towards and pressure on the press."

https://rsf.org/en/classement/2026/americas -> also explains how Trump's presidency has led to one of the most severe crises for press freedom in modern US history.

**Organized Crime: Links issue of organized crime to Ecuador, whose press freedom score, not global rank, dropped the most out of countries in the Americas since last year, along with Peru, which is also in the "very serious" category alongside Venezuela, Cuba and Nicaragua.**

"In countries where **organised crime** kills, Index rankings fall drastically. This is the case in Ecuador (125th), which has dropped 31 places since Darwin Baque and Patricio Aguilar were murdered in 2025. The same year, Peru (144th, -14) was impacted by the murder of four journalists. Guarantees for press freedom in Venezuela (159th) remain deeply uncertain despite the release of detained journalists earlier in the year. Finally, Cuba (160th) is undergoing a profound crisis that is forcing the few remaining independent journalists to operate underground, and Nicaragua’s (168th) media landscape lies in ruins, characterised by systematic repression and a lasting collapse in working conditions for journalists."
